Feature Engineering

```python
df
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df[['Sex']])
df['ChestPainType']=le.fit_transform(df[['ChestPainType']])
df['RestingECG']=le.fit_transform(df[['RestingECG']])
df['ExerciseAngina']=le.fit_transform(df[['ExerciseAngina']])
```

```python
df.head()
```

```python
from sklearn.preprocessing  import MinMaxScaler, StandardScaler
```

```python
mms=MinMaxScaler()
stc=StandardScaler()
```

```python
df['Age']=mms.fit_transform(df[['Age']])
df['Cholesterol']=stc.fit_transform(df[['Cholesterol']])
df['MaxHR']=stc.fit_transform(df[['MaxHR']])
df['RestingBP']=stc.fit_transform(df[['RestingBP']])
df['Oldpeak']=stc.fit_transform(df[['Oldpeak']])
```

```python
df['ST_Slope']=le.fit_transform(df[['ST_Slope']])
```

Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df[df.columns.drop(['HeartDisease','RestingBP','RestingECG'])].values
y=df['HeartDisease'].values
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=2024)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
ypred_lr=lr.predict(xtest)
```

```python
lr.score(xtest,ytest)
```

```python
lr.score(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report
```

```python
a=classification_report(ypred_lr,ytest)
```

```python
print(a)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(criterion='entropy',max_depth=8,max_leaf_nodes=30)
dtc.fit(xtrain,ytrain)
ypred_dtc=dtc.predict(xtest)
```

```python
dtc.score(xtest,ytest)
```

```python
dtc.score(xtrain,ytrain)
```

```python
b=classification_report(ypred_dtc,ytest)
```

```python
print(b)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(criterion='entropy',max_depth=6,max_leaf_nodes=38)
rfc.fit(xtrain,ytrain)
ypred_rfc=rfc.predict(xtest)
```

```python
rfc.score(xtest,ytest)
```

```python
rfc.score(xtrain,ytrain)
```

```python
c=classification_report(ypred_rfc,ytest)
```

```python
print(c)
```
